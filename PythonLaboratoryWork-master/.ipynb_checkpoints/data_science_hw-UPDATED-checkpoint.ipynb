## 3. Накачка и сброс (1 балл)
Криптовалютные биржи до сих пор остаются маргинальным местом, эдаким диким западом финансового мира. Как следствие, здесь процветают схемы относительно
честного отъема денег. Одна из них - pump'n'dump (накачка и сброс). Она выглядит следуюшим образом. Несколько крупных игроков или много мелких договариваются
вместе купить малоизвестную монету с низкой ценой и объемом торгов. Это приводит к мгновенному взлету цены (pump), далее приходят неопытные игроки в надежде
успеть заработать на таком росте. В этот момент организаторы схемы начнают все продавать (dump). Весь процесс занимает от нескольких минут до нескольких часов.

#### Ваша задача найти самый сильный pump'n'dump монеты на заданном промежутке времени. Для этого для каждого дня определим число pnd равное отношению максимальной цены монеты в данный день к максимуму из цен открытия и закрытия в тот же день. Нужно найти день когда pnd был максимален и величину pnd.

```python
def find_most_severe_pump_and_dump(coins, symbol, start_date, end_date):
    requested_data = coins[(coins.symbol == symbol) & (start_date <= coins.dt) & (coins.dt <= end_date)][['dt', 'high', 'open', 'close']]
    requested_data['pnd'] = np.where(requested_data.open >= requested_data.close, requested_data.high / requested_data.open, requested_data.high / requested_data.close)
    ans = np.array(requested_data[(requested_data.pnd == requested_data.pnd.max())][['pnd', 'dt']])
    return ans[0][0], ans[0][1]
```

```python
find_most_severe_pump_and_dump(coins, symbol='BTC', start_date='2017-06-01', end_date='2018-06-01')
```

#### Сравните эти значения для разных монет.

```python
from datetime import datetime

start_date = '1970-01-01'
end_date = datetime.now().strftime('%Y-%m-%d')

values = []

for symbol in set(coins['symbol']):
    values.append((find_most_severe_pump_and_dump(coins, symbol=symbol, start_date=start_date, end_date=end_date)[0], symbol))
    
print('Currency : Most severe pump and dump')
    
for elem in sorted(values):
    print(elem[1], ':', elem[0])
```

Вывод: pump'n'dump достаточно распространены. Чем меньше суммарная капитализация монеты, тем легче ее pump'n'dump'нуть, поэтому с тем же биткоином пампы были незначительнее, чем с более дешевыми и меньшими в капитализации монетами

## 4. Окупаемость инвестиций (1 балл)

#### Вам нужно посчитать окупаемость инвестиций в криптовалюты на заданном промежутке времени. Окупаемость определяется как отношение изменения цены портфеля к исходной цене портфеля. Цена портфеля - это суммарная стоимость (в USD) всех монет в портфеле.
investments - dict в котором ключи - это названия монет, значения - это сумма вложений в эту монету (в USD)

```python
def compute_roi(coins, investments, start_date, end_date):
    starting_price, final_price = 0, 0
    for currency in investments:
        starting_currency_price = int(coins[(coins['symbol'] == currency) & (coins['dt'] == start_date)]['price'])
        final_currency_price = int(coins[(coins['symbol'] == currency) & (coins['dt'] == end_date)]['price'])
        starting_price += starting_currency_price * investments[currency]
        final_price += final_currency_price * investments[currency]
    return float(final_price) / starting_price
```

## 5. Технический анализ (1 балл)

Технический анализ это способ предсказания поведения графика по некоторым вспомогательным величинам построенным по исходному графику. Один из простейших методов технического анализа - границы Болинджера. Кто-то верит, что график касаясь границы от него должен отражаться.

#### Нарисуйте график цены, скользящее среднее и [границы Боллинджера](https://en.wikipedia.org/wiki/Bollinger_Bands) c параметрами N (window) = 21, K (width) = 2.

Границы считаются очень просто: (MA + Kσ) и (MA - Kσ), где MA - скользящее среднее за N дней, а σ - скользящее стандартное отклонение за N дней.

Тут вам поможет функция rolling для подсчёта среднего и стандартного отклонения по скользящему окну.

Не забудьте подписать график и оси, отрисовать легенду и выбрать для нее лучшее расположение.

```python
def plot_bollinger_bands(coins, symbol, window, width):
    symbol_data = coins[(coins['symbol'] == symbol)]
    symbol_data.loc[:, 'MA'] = symbol_data['price'].rolling(window=window).mean()
    symbol_data.loc[:, 'std'] = symbol_data['price'].rolling(window=window).std()
    symbol_data = symbol_data[['dt', 'price', 'MA', 'std']]
    symbol_data.dt = pd.to_datetime(symbol_data.dt)
    symbol_data.dt = symbol_data.dt.apply(mdates.date2num)
    fig = plt.figure(figsize=(16, 9))
    ax = fig.gca()
    ax.xaxis_date()
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.plot(symbol_data['dt'].values, symbol_data['price'].values, label='Our coin', color='g')
    plt.plot(symbol_data['dt'].values, symbol_data['MA'].values + width * symbol_data['std'], label='Top border', color='r')
    plt.plot(symbol_data['dt'].values, symbol_data['MA'].values - width * symbol_data['std'], label='Bottom border', color='b')
    plt.title('Theory test')
    plt.xlabel('date')
    plt.ylabel('price')
    plt.grid(True)
    plt.legend(loc='best')
```

```python
plot_bollinger_bands(coins=coins, symbol='EOS', window=21, width=2)  # тут должен появиться график
```

## 6. Капитализация как индикатор (1 балл)

Многие люди, которые торгуют криптовалютой, любят смотреть на капитализацию. Давайте поймём почему.

#### Нарисуйте еще два графика. На первом должна быть общая капитализация биткойна (BTC), эфира (ETH), еос (EOS), биткойн кэша (BCH), стеллара (XLM) и лайткойна (LTC). На втором - доли капитализаций этих монет от общей капитализации рынка. При этом используйте данные начиная с 2017-07-01.

```python
from datetime import datetime, timedelta

def plot_coins_capitalizations(coins, symbols, start_date):
    currency_capitalization = []
    date_calculated = False
    dates = []
    for idx, currency in enumerate(symbols):
        current_currency_capitalization = [0]
        current_currency_data = coins[(coins['symbol'] == currency) & (coins['dt'] >= start_date)]
        dates.append([])
        current_currency_data.dt = pd.to_datetime(current_currency_data.dt)
        current_currency_data.dt = current_currency_data.dt.apply(mdates.date2num)
        for index, row in current_currency_data.iterrows():
            dates[idx].append(row['dt'])
            current_currency_capitalization.append(current_currency_capitalization[-1] + row['market'])
        currency_capitalization.append(current_currency_capitalization[1::])
    
    capitalization_data = coins[coins['dt'] >= start_date]
    capitalization_data.dt = pd.to_datetime(capitalization_data.dt)
    capitalization_data.dt = capitalization_data.dt.apply(mdates.date2num)
    grouper = capitalization_data.groupby(['dt'])
    group = grouper['market'].agg(sum)
    dates.append(list(group.index.get_level_values(0)))
    currency_capitalization.append(list(grouper['market'].agg(sum)))
    
    for i in range(1, len(currency_capitalization[-1])):
        currency_capitalization[-1][i] += currency_capitalization[-1][i - 1]
        
    symbols = list(symbols)
    symbols.append('Total market')
    
    good_dates = set(dates[0]) # получамем список дат, встречающихся в датах торгов каждой валюты
    
    for i in range(1, len(dates)):
        good_dates = good_dates.intersection(set(dates[i]))
    
    
    relative_capitalization = []
    for i in range(len(symbols)):
        relative_capitalization.append([])
        for idx, date in enumerate(dates[i]):
            if date in good_dates:
                relative_capitalization[i].append(currency_capitalization[i][idx])
    
    fig = plt.figure(figsize=(16,9))
    ax = fig.gca()
    ax.xaxis_date()
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    
    good_dates = sorted(list(good_dates))
    
    for i in range(len(symbols)):
        ax.plot(good_dates, relative_capitalization[i], label=symbols[i])
    plt.xlabel('date')
    plt.ylabel('capitalization')
    plt.title('Total capitalizaion')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    for i in range(len(symbols) - 1):
        for l in range(len(relative_capitalization[i])):
            if relative_capitalization[-1][l] != 0:
                relative_capitalization[i][l] = float(relative_capitalization[i][l])/float(relative_capitalization[-1][l])
    
    for i in range(len(relative_capitalization[-1])):
        relative_capitalization[-1][i] = 1
    
    fig = plt.figure(figsize=(16,9))
    ax = fig.gca()
    ax.xaxis_date()
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    
    for i in range(len(symbols)):
        ax.plot(good_dates, relative_capitalization[i], label=symbols[i])
    plt.xlabel('date')
    plt.ylabel('capitalization part')
    plt.title('Relative capitalizaion')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_coins_capitalizations(
    coins=coins,
    symbols=('BTC', 'ETH', 'EOS', 'BCH', 'XLM', 'LTC'),
    start_date='2017-07-01'
)
```

## 7. Корреляции монет (1 балл)

#### Теперь нужно подробнее посмотреть на корреляции средних капитализаций монет. При этом будем смотреть на среднее сглаженное за последние window дней до дня date с коэффициентом сглаживания alpha для набора монет symbols.  
#### Реализуйте функцию, которая будет возвращать квадратный DataFrame с числом строк и столбцов равным числу рассматриваемых монет и со значениями корреляций.

```python
def calc_coins_correlations(coins, date, symbols, window, alpha):
    finish_date = date
    date = list(map(int, date.split('-')))
    date = datetime(date[0], date[1], date[2], 0, 0, 0) - timedelta(days=window)
    start_date = date.strftime('%Y-%m-%d')
    temp_frame = coins[(coins['dt'] >= start_date) & (coins['dt'] <= finish_date)][['symbol', 'market']]
    values = []
    for symbol in symbols:
        values.append(np.array(temp_frame[temp_frame['symbol'] == symbol].ewm(alpha=alpha, adjust=False).mean()['market']))
    values = np.rot90(values)
    frame = pd.DataFrame.from_records(values, columns=symbols)
    return frame.corr()
```

```python
correlations = calc_coins_correlations(coins, date="2018-06-06",
                                       symbols=['BTC', 'ETH', 'EOS', 'BCH', 'XLM', 'LTC', 'ADA'],
                                       window=21, alpha=0.1)
# Теперь посмотрим на эти корреляции следующим образом:
correlations.style.background_gradient(cmap='coolwarm').set_precision(2)
```

## 8. Анализ одной стратегии (2 балла)

Разберем один мечтательный пример. Посмотрим какую прибыль могла бы нам принести хрестоматийная торговая стратегия основанная на скользящих средних. Стратегия выглядит следующим образом: мы строим две скользящие среднии для графика цены. С маленьким окном (ведущее скользящее среднее) и с бОльшим окном (запаздывающее скользящее среднее). Мы покупаем, когда ведущее среднее становится больше запаздывающего, и продаем в противном случае. Посмотрим на пример

```python
def plot_moving_averages(coins, symbol, leading_window, lagging_window, start_date, end_date):
    coin = coins[coins['symbol'] == symbol][start_date:end_date]
    price = coin['price']
    leading_mean = price.rolling(window=leading_window).mean()
    lagging_mean = price.rolling(window=lagging_window).mean()
    
    fig = plt.figure(figsize=(16, 9))
    ax = fig.add_subplot(111)
    
    ax.set_title('Price action for {}'.format(symbol))
    ax.plot(leading_mean, color='green', label='MA{}'.format(leading_window))
    ax.plot(lagging_mean, color='red', label='MA{}'.format(lagging_window))
    ax.plot(price, color='blue', label='price')
    ax.set_xlabel('Date')
    ax.set_ylabel('Price')
    ax.legend(loc='best')
    ax.grid(True)
    plt.show() 

plot_moving_averages(
    coins=coins, 
    symbol='BTC', 
    leading_window=21, 
    lagging_window=50, 
    start_date='2017-05-01',
    end_date='2018-08-01')
```

Видно, что для скользящее среднее с бОльшим окном медленнее реагирует на изменение цены. Именно на этой идее и основана торговая стратегия.
#### Реализуйте функцию, которая строит два графика. На правом будут изображены цена и скользящие средние. На левом - во сколько раз изменится размер вложений при использовании нашей стратегии и при обычном инвестировании 

#### Notes:
Давайте использовать только цены закрытия. При этом, чтобы узнать цены за вчерашний день, стоит использовать метод shift(1) у Series.
Отношение цен закрытия за сегодня и за вчера - это мой multiplier за сегодняшний день. При этом давайте строить графики накопления для multipliers. Т.е. если мы смотрим на 3 дня и в первый день multiplier = 1.5, во второй- 0.5 и в третий 2. То график будет выглядеть так: (1.5, 1.5 * 0.5, 1.5 * 0.5 * 2).

При использовании нашей новой стратегии мы будем либо покупать, если ведущее среднее становится больше запаздующего на некоторый threshold (при этом лучше разницу сперва поделить на цену), либо оставлять всё как есть. При этом, конечно, нужно, принимая решения за сегодняшний день, смотреть только на статистику из прошлого.

```python
def assign_status(row, threshold):
    if row['leading'] - row['lagging'] >= threshold:
        return 'open'
    else:
        return 'close'


def assign_multiplier(row):
    if row['status'] == 'close':
        return row['prev_mult']
    else:
        return row['mult']

    
def plot_moving_averages_strategy(
    coins, symbol, lead_window, lag_window, threshold, start_date, end_date
):
    start_date, end_date = str(start_date)[:10], str(end_date)[:10]
    investment_start_date = list(map(int, start_date.split('-')))
    investment_start_date = datetime(investment_start_date[0], investment_start_date[1], investment_start_date[2], 0, 0, 0)
    investment_start_date = investment_start_date + timedelta(days=max(lead_window, lag_window))
    temp_df = coins[(coins['symbol'] == symbol) & (coins['dt'] >= start_date) & (coins['dt'] <= end_date)]
    temp_df.loc[:,'mult'] = (temp_df['close'] / temp_df['close'].shift(1))
    temp_df['mult'].fillna(1, inplace=True)
    temp_df.loc[:, 'total_mult'] = temp_df['mult'].cumprod()
    temp_df.loc[:, 'leading'] = temp_df['price'].rolling(window=lead_window).mean()
    temp_df.loc[:, 'lagging'] = temp_df['price'].rolling(window=lag_window).mean()
    temp_df['lagging'].fillna(1, inplace=True)
    temp_df['leading'].fillna(2 + threshold, inplace=True) # чтобы купить с самого начала
    temp_df.loc[:, 'status'] = temp_df.apply(lambda row: assign_status(row, threshold), axis=1)
    
    coin = coins[coins['symbol'] == symbol][start_date: end_date]
    price = coin['price']
    leading_mean = price.rolling(window=lead_window).mean()
    lagging_mean = price.rolling(window=lag_window).mean()
    
    fig = plt.figure(figsize=(16, 9))
    ax = fig.add_subplot(121)
    
    ax.set_title('Price action for {}'.format(symbol))
    ax.plot(leading_mean, color='green', label='MA{}'.format(lead_window))
    ax.plot(lagging_mean, color='red', label='MA{}'.format(lag_window))
    ax.plot(price, color='blue', label='price')
    ax.set_xlabel('Date')
    plt.xticks(rotation=45)
    ax.set_ylabel('Price')
    ax.legend(loc='best')
    ax.grid(True)
    
    ax = fig.add_subplot(122)
    ax.plot(temp_df['total_mult'], label='Investment')
    ax.plot(temp_df[temp_df['status'] == 'open']['mult'].cumprod(), label='Speculation')
    ax.set_xlabel('Date')
    plt.xticks(rotation=45)
    ax.set_ylabel('Multiplier')
    ax.set_title('Multiplier action for {}'.format(symbol))
    ax.legend(loc='best')
    ax.grid(True)
    plt.show()
    return
```

```python
# Теперь на основе реализованной функции сделаем интерактивные графики и поизучаем, что получилось:
symbol_selector = ipywidgets.Dropdown(
    options=('BTC', 'ETH', 'EOS', 'BCH', 'XLM', 'LTC', 'ADA'),
    index=0,
    value='BTC',
    layout={'width': '700px'},
    continuous_update=False
)

lead_window_slider = ipywidgets.IntSlider(
    value=21, 
    min=1, 
    max=200, 
    step=1,
    layout={'width': '700px'},
    continuous_update=False)

lag_window_slider = ipywidgets.IntSlider(
    value=50,
    min=1, 
    max=200, 
    layout={'width': '700px'},
    step=1, continuous_update=False)

threshold_slider = ipywidgets.FloatSlider(
    min=0, 
    max=0.20, 
    step=0.001, 
    value=0.025,
    layout={'width': '700px'},
    continuous_update=False)

start_date_slider = ipywidgets.SelectionSlider(
    options=pd.date_range('2013-04-28', '2018-06-06', freq='D'),
    index=0,
    value=pd.Timestamp('2017-05-01'),
    layout={'width': '700px'},
    continuous_update=False
)

end_date_slider = ipywidgets.SelectionSlider(
    options=pd.date_range('2013-04-28', '2018-06-06', freq='D'),
    index=0,
    value=pd.Timestamp('2018-01-01'),
    layout={'width': '700px'},
    continuous_update=False
)

ipywidgets.interact(
    plot_moving_averages_strategy,
    coins=ipywidgets.fixed(coins),
    symbol=symbol_selector,
    lead_window=lead_window_slider,
    lag_window=lag_window_slider,
    threshold=threshold_slider,
    start_date=start_date_slider,
    end_date=end_date_slider
)
```
